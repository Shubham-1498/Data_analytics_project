Checked for presence of outliers 

```python
# Created copy of dataframe 
Hr1 = Hr.copy()
```

```python
Hr1
```

```python
from scipy import stats

# Calculated the Z-scores for each data point in the column
z_scores = stats.zscore(Hr1['Age'])

# Define a threshold for considering a Z-score as an outlier (e.g., Z-score greater than 3 or less than -3)
threshold = 3

# Find the indices of outliers based on the threshold
outlier_indices = (z_scores) > threshold | (z_scores<-threshold)

outlier_indices
```

```python
Hr1[outlier_indices==True]
```

#  firstly separated the numerical & non numerical columns 

```python
numerical_columns = Hr1.select_dtypes(include =['number']).columns.tolist()

categorical_columns = Hr1.select_dtypes(exclude = ['number']).columns.tolist()

print(numerical_columns)

print(categorical_columns)
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Assuming Hr1 is your DataFrame and numerical_columns is a list of numerical column names
fig, axes = plt.subplots(nrows=len(numerical_columns), ncols=1, figsize=(10, 6 * len(numerical_columns)))

# Iterate over numerical columns
for i, col in enumerate(numerical_columns):
    # Plot boxplot for each column
    sns.boxplot(x=Hr1[col], ax=axes[i])
    
    # Calculate Z-scores and remove outliers
    z_scores = stats.zscore(Hr1[col])
    threshold = 3
    outlier_indices = (z_scores > threshold) | (z_scores < -threshold)
    Hr2 = Hr1[~outlier_indices]  # Filtering DataFrame to remove outliers

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
